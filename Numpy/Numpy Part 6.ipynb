```python
import numpy as np
import matplotlib.pyplot as plt
```

### Trignometric Function using numpy

```python
pi=np.pi
```

```python
np.sin(pi/2)
```

```python
np.cos(2*pi)
```

```python
np.tan(pi/4)
```

### using Matplotlib 

```python
x=np.arange(1,10)
y=np.arange(1,10)

plt.figure(figsize=(6,6))
plt.plot(x,y,"b.")
plt.show()
```

```python
x=np.arange(0,2*pi,0.1)
y=np.sin(x)
print(y)

plt.title("Sin curve")
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,2*pi,0.1)
y=np.cos(x)
print(y)

plt.title("Cos curve")
plt.plot(x,y)
plt.show()
```
