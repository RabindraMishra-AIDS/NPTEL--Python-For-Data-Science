### Fill null values 

```python
df.fillna(2,inplace=True)
df.isnull().sum().sum()
```

```python
df2=df1
df1=df1.fillna({"Content":"none","Headquarters":"Banglore"}) #Filling null values for separate column
df1.isnull().sum()
```

### Forward Fill ffill ; Backward Fill 

```python
df2.fillna(method="ffill") #Previous row non-null value will get filled 
```

```python
df2.fillna(method="bfill",axis=1) #After value of col (axis=1) will get filled 
```

### Filling using mean,sum

```python
df['Units'].fillna(value=df['Units'].mean())
```
