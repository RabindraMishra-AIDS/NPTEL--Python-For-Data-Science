```python
import pandas as pd
import numpy as np
```

### Replace Function in Pandas

```python
df=pd.read_csv(r"C:\Users\admin\Desktop\sample2.csv")
df.head(3)
```

#### shift+tab to acess docstring

```python
df.replace(to_replace=23,value=45).head(3)
```

### Replacing Large Set of Numbers

```python
df.replace(to_replace=[25,56,78],value=1).head(1)
```

```python
df.replace(to_replace=[25,56,78],value=[3,4,7]).head(3) #Replacing coressponding Values
```

### Replacing a Particular Column

```python
df['Branch'].replace(to_replace='CS',value="CSE").head(3)
```

### For replacing as Regular expression need to set regex as true

```python
df.replace('[A-Za-z]',0 ,regex=True)
```

### Can use ffill & bfill to replace a value

```python
df.replace(to_replace=24,method="ffill").head(3) #Number previous to 24 will replace it
```
